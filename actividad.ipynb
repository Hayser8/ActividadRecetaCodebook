## **Resumen de Metadatos**

```python
uniques = {}
for col in df.columns:
    try:
        uniques[col] = df[col].nunique(dropna=False)
    except TypeError:
        uniques[col] = df[col].dropna().astype(str).nunique()

metadata = pd.DataFrame({
    'dtype':  df.dtypes,
    'nulos':  df.isnull().sum(),
    'Ãºnicos': pd.Series(uniques)
})
metadata
```

```python
from pandas.api.types import is_numeric_dtype, is_datetime64_any_dtype
# 2. Definir funciÃ³n para generar el codebook
def generate_codebook(df):
    records = []
    for col in df.columns:
        series = df[col]
        dtype = series.dtype
        # contar faltantes
        missing_count = series.isnull().sum()
        # valores posibles (si hay <=20 Ãºnicos los listamos, si no mostramos el conteo)
        uniques = series.dropna().unique()
        if len(uniques) <= 20:
            posibles = uniques.tolist()
        else:
            posibles = f'{len(uniques)} valores Ãºnicos'
        # estadÃ­sitcas bÃ¡sicas
        if is_numeric_dtype(series):
            stats = series.describe().to_dict()
        elif is_datetime64_any_dtype(series):
            stats = {'min': series.min(), 'max': series.max()}
        else:
            top = series.mode().iloc[0] if not series.mode().empty else None
            freq = series.value_counts().iloc[0] if not series.value_counts().empty else None
            stats = {'unique': series.nunique(), 'top': top, 'freq': freq}
        records.append({
            'nombre': col,
            'descripcion': '',                      # completar manualmente
            'tipo de dato': str(dtype),
            'unidades': '',                         # completar si aplica
            'valores posibles': posibles,
            'valores especiales/faltantes': missing_count,
            'estadisticas basicas': stats,
            'fuente original': '',                  # completar manualmente
            'transformaciones': '',                 # documentar aquÃ­
            'notas': ''                             # anotaciones adicionales
        })
    return pd.DataFrame(records)

# 3. Crear y visualizar el codebook
codebook = generate_codebook(df)
codebook.head()

# 4. (Opcional) Exportar a CSV o Excel para revisiÃ³n externa
codebook.to_csv('codebook.csv', index=False)
# o bien
codebook.to_excel('codebook.xlsx', index=False)
```
